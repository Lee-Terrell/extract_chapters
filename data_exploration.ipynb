```python
# Import dependencies
import sys
import os
import json
import subprocess
import re
from pathlib import Path
from pydub import AudioSegment
```

```python
audio_path = "resources/AdventuresOfGrandfatherFrog_librivox.m4b"
```

```python

# Use subprocess to execute ffprobe in order to get chapter data
result = subprocess.run(
            ['ffprobe', '-v', 'quiet', '-print_format', 'json', '-show_chapters', audio_path],
            capture_output=True,
            text=True,
            check=True)
```

```python
# Save chapter data as a python dictionary
data = json.loads(result.stdout)
```

```python
# Convert JSON string to python dictionary of chapters
json.loads(result.stdout)
```

```python
# Use subprocess to execute ffprobe in order to get metadata
metadata_mp4 = subprocess.run(
            ['ffprobe', '-print_format', 'json', '-show_format', audio_path],
            capture_output=True,
            text=True,
            check=True)
```

```python
# Convert JSON string to python dictionary and view
json.loads(metadata_mp4.stdout)
```
